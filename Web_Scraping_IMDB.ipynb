```
from urllib.request import urlopen
from requests import get
from bs4 import  BeautifulSoup
```

```
url = "https://www.imdb.com/list/ls033632731/"
response = get(url)
response.text
```

```
soup = BeautifulSoup(response.text,"html.parser")
soup.find_all("div",class_ ="lister-item-content")
```

```
soup = BeautifulSoup(response.text,"html.parser")
num = BeautifulSoup(str(soup.find_all("span",class_= "lister-item-index unbold text-primary"))).get_text()
title = BeautifulSoup(str(soup.find_all('a', href=True))).get_text()
year = BeautifulSoup(str(soup.find_all("span",class_= "lister-item-year text-muted unbold"))).get_text()
duration = BeautifulSoup(str(soup.find_all("span",class_= "runtime"))).get_text()
genre = BeautifulSoup(str(soup.find_all("span",class_= "genre"))).get_text()
ratings = BeautifulSoup(str(soup.find_all("span",class_= "ipl-rating-star__rating")[::23])).get_text()
description = BeautifulSoup(str(soup.find_all("p",class_=""))).get_text()
cast = BeautifulSoup(str(soup.find_all("p",class_="text-muted text-small"))).get_text()
```

```
len(soup.find_all("span",class_= "runtime"))
```

```
import pandas as pd
import re
imdb_telugutop100movies = pd.DataFrame(columns=["Rank","Title","Year_Released","Genre","Rating","Movie_Description","Director","Cast&Crew"]) 
imdb_telugutop100movies[imdb_telugutop100movies.columns[0]] = pd.DataFrame(pd.DataFrame([num.strip("[ ]")])[0].str.split(",",expand = True).transpose()[0].str.strip("."))[0]
imdb_telugutop100movies[imdb_telugutop100movies.columns[1]] = pd.DataFrame(pd.DataFrame(re.findall(r'(  \n)(.*?)(  \n\n\n\n)',title,re.DOTALL))[1].str.strip(", ,"))[1]
imdb_telugutop100movies[imdb_telugutop100movies.columns[2]] = pd.DataFrame(pd.DataFrame([year.strip("[ ]")])[0].str.split(",",expand = True).transpose()[0].str.strip("( )"))[0]
# imdb_telugutop100movies[imdb_telugutop100movies.columns[3]] = pd.DataFrame([duration.strip("[ ]")])[0].str.split(",",expand = True).transpose()[0]
imdb_telugutop100movies[imdb_telugutop100movies.columns[3]] = pd.DataFrame(pd.DataFrame([genre.strip("[ ]")])[0].str.split(", \n",expand = True).transpose()[0].str.strip("\n"))[0]
imdb_telugutop100movies[imdb_telugutop100movies.columns[4]] = pd.DataFrame([ratings.strip("[ ]")])[0].str.split(",",expand = True).transpose()[0]
imdb_telugutop100movies[imdb_telugutop100movies.columns[5]] = pd.DataFrame(pd.DataFrame([description.strip("[ ]")])[0].str.split(", \n",expand = True).transpose()[1:][0].str.strip("\n")).reset_index(drop = True)[0]
imdb_telugutop100movies[imdb_telugutop100movies.columns[6]] = pd.DataFrame(pd.DataFrame(re.findall(r'Directo(r|s:)(.*?)(Stars:)',cast.strip("[ ]"),re.DOTALL))[1].str.strip(":\n \n| \n"))[1]
imdb_telugutop100movies[imdb_telugutop100movies.columns[7]] = pd.DataFrame(([pd.DataFrame(pd.DataFrame(re.findall(r'(Stars:)(.*?)(\nVotes:)',cast.strip("[ ]"),re.DOTALL))[1])[1][i].replace("\n","") for i in range(100)]))[0]
```
